```python
import numpy as np
import matplotlib.pyplot as plt
from mapio.gridcontainer import GridHDFContainer
import os.path
```

```python
stackfile = 'losangeles_shakestack.hdf'
container = GridHDFContainer.load(stackfile)
print(container.getArrays())
pgacube,metadata = container.getArray('pga')
```

```python
pgacube.shape
```

```python
pgamean = np.nanmean(pgacube,axis=2);
mean_mean = pgamean.mean()
mean_std = pgamean.std()
vmin = 0
vmax = mean_mean + mean_std*3
plt.imshow(pgamean,vmin=vmin,vmax=vmax);
plt.colorbar();
```

```python
pgamax = np.nanmax(pgacube,axis=2);
max_mean = pgamax.mean()
max_std = pgamax.std()
vmin = 0
vmax = max_mean + max_std*3
plt.imshow(pgamax,vmin=vmin,vmax=vmax);
plt.colorbar();
```

```python
pgamedian = np.nanmedian(pgacube,axis=2);
med_mean = pgamedian.mean()
med_std = pgamedian.std()
vmin = 0
vmax = med_mean + med_std*3
plt.imshow(pgamedian,vmin=vmin,vmax=vmax);
plt.colorbar();
```

```python
pga_array = np.sort(pgacube[35,30,:]);
plt.plot(pga_array,'o');
```
